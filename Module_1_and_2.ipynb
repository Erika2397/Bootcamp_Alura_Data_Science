````
# Importando a biblioteca de DateTime
from datetime import date

date.today()
````

````
# Chamando o mensal.index
mensal.index
````

````
# Criando uma função para transformar o mensal.index em DateTime
meses = {
    'Jan': 1, 
    'Fev': 2,
    'Mar': 3,
    'Abr': 4,
    'Mai': 5,
    'Jun': 6,
    'Jul': 7,
    'Ago': 8,
    'Set': 9,
    'Out':10,
    'Nov':11,
    'Dez':12
}

def para_dia(ano_mes: str):
  ano:int = int(ano_mes[:4])
  mes:str = ano_mes[5:]
  mes_numero:int = meses[mes]
  return date(ano, mes_numero, 1)
  
mensal.index.map(para_dia)
````

````
# Trocando o index atual pelo novo index com DataTime
mensal.index= mensal.index.map(para_dia)
mensal.head()
````

````
# Plotando novamente o gráfico sem Index.Locator
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index, y='São Paulo')
plt.ylim(0,350)
plt.grid(True)
plt.xticks(rotation=25)
plt.show()
````
